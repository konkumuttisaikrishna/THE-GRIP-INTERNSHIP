```python
matches.umpire2.fillna('C Shamshuddin',inplace=True)
matches['umpire1'].fillna(matches['umpire1'].mode()[0],inplace=True)
matches.city.fillna('unknown',inplace=True)
matches.winner.fillna('unknown',inplace=True)
matches.player_of_match.fillna('unknown',inplace=True)
```

```python
matches.isnull().sum()
```

```python
matches['winner'] = matches['winner'].apply(lambda x:x.replace('Rising Pune Supergiant',"Rising Pune Supergiants"))
matches['winner'] = matches['winner'].apply(lambda x:x.replace('Delhi Daredevils',"Delhi Capitals"))
matches['winner'] = matches['winner'].apply(lambda x:x.replace('Rising Pune Supergiantss',"Rising Pune Supergiants"))

```

```python
matches['toss_winner'] = matches['toss_winner'].apply(lambda x:x.replace('Rising Pune Supergiant',"Rising Pune Supergiants"))
matches['toss_winner'] = matches['toss_winner'].apply(lambda x:x.replace('Delhi Daredevils',"Delhi Capitals"))
matches['toss_winner'] = matches['toss_winner'].apply(lambda x:x.replace('Rising Pune Supergiantss',"Rising Pune Supergiants"))
```

```python
matches['city'] = matches['city'].apply(lambda x:x.replace("Bengaluru","Bangalore"))
```

```python
matches['city'].value_counts()
```

```python
matches.head()
```

## Data Analysis on IPL

### MOST OF THE WINNING TEAM IN IPL

```python
plt.figure(figsize=(35,10))
sns.countplot(x='winner',data=matches)
plt.xticks(rotation=90,fontsize=20)
plt.title('MOST OF THE WINNING TEAM IN IPL')
plt.show()
```

- Mumbai Indians and Chennai Super Kings wins most of the ipl matches.

### NO OF TEAMS PARTICIPATED IN EACH SEASON

```python
plt.figure(figsize=(35,10))
sns.countplot(x='season',data=matches)
plt.xticks(rotation=90,fontsize=20)
plt.title('NO OF TEAMS PARTICIPATED IN EACH SEASON')
plt.show()
```

- 2013 TO 2011 MORE TEAMS PLAYS IPL.

### TEAM WHICH WINS MORE IN TOSS

```python
plt.figure(figsize=(35,10))
sns.countplot(x='toss_winner',data=matches)
plt.xticks(rotation=90,fontsize=20)
plt.title('TEAM WHICH WINS MORE IN TOSS')
plt.show()
```

- MUMBAI INDIANS WINS MORE IN THE TOSS.

### Does teams choosed to bat or field first, after winning toss?

```python
toss=matches.toss_decision.value_counts()
labels=np.array(toss.index)
sizes = toss.values
colors = ['purple', 'brown']
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)

plt.title('Decision taken by captains after winning tosses')
plt.axis('equal')
plt.show()
```

- MOST OF TEAM CHOOSES FIELD AFTER WINNING THE TOSS.

- IN MUMBAI, BANGALORE, KOLKATA AND DELHI HOSTED HIGHEST NO OF MATCHES.

### BATING ANALYSIS

### BATSMEN WITH HIGHEST NO  OF THE RUNS

```python
batsman_df = delevery.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs',ascending=False)
```

```python
batsman=batsman_df.head(10)
```

```python
batsman
```

```python
plt.figure(figsize=(40,20))
sns.barplot(x='batsman',y='batsman_runs',data=batsman)
plt.title('PLAYERS WITH HIGHEST NO OF RUNS')
plt.xticks(rotation=90,fontsize=30)
plt.show()
```

- VIRAT KOHLI AND SURESH RAINA HAS HIGHEST NO OF RUNS.
- ROHIT SHARMA AND DAVID WARNER OCCUPIES 3RD AND 4TH POSTION.

### BATSMEN WITH HIGHEST NO OF SIXES.

```python
batsman_sixes = delevery.groupby('batsman')['batsman_runs'].agg(lambda x: (x==6).sum()).reset_index().sort_values(by='batsman_runs',ascending=False)
```

```python
sixes=batsman_sixes.head(20)
```

```python
plt.figure(figsize=(40,20))
sns.barplot(x='batsman',y='batsman_runs',data=sixes)
plt.title('PLAYERS WITH HIGHEST NO OF SIXES')
plt.xticks(rotation=90,fontsize=30)
plt.show()
```

- CH GAYLE HAS HIGHEST NO OF SIXES.
- AB DE VILLIERS AND MS DHONI OCCUPIES 2ND AND 3RD POSTION IN HIGHEST NO OF SIXES.

### BATSMEN WITH HIGHEST NO OF FOURS.

```python
batsman_fours = delevery.groupby('batsman')['batsman_runs'].agg(lambda x: (x==4).sum()).reset_index().sort_values(by='batsman_runs',ascending=False)
```

```python
fours=batsman_fours.head(20)
```

```python
plt.figure(figsize=(40,20))
sns.barplot(x='batsman',y='batsman_runs',data=fours)
plt.xticks(rotation=90,fontsize=30)
plt.title('PLAYERS WITH HIGHEST NO OF FOURS')
plt.show()
```

- S DHAWAN AND SURESH RAINA HAS HIGHEST NO OF FOURS.

### BOWLING ANALYSIS

### MORE NO OF WIDES BY THE BOWLER

```python
bowler_wides = delevery.groupby('bowler')['wide_runs'].agg('sum').reset_index().sort_values(by='wide_runs',ascending=False)
```

```python
plt.figure(figsize=(40,20))
sns.barplot(x='bowler',y='wide_runs',data=wides)
plt.xticks(rotation=90,fontsize=30)
plt.title('BOWLER WITH NO OF WIDES')
plt.show()
```

### HIGHEST NO OF WICKETS BY THE BOWLER.
