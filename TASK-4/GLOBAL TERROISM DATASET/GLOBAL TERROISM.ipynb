### Cleaning the data

```python
terr = terr[['year','month','day','Country','State','Region','City','Latitude','Longitude','Attack_Type','Killed','Wounded','Group','Target','Weapon_Type','Target_type']]
```

```python
terr.head()
```

```python
terr.info()
```

```python
terr.isnull().sum()
```

```python
terr['Wounded'] = terr['Wounded'].fillna(0)
terr['Killed'] = terr['Killed'].fillna(0)
```

```python
terr.isnull().sum()
```

```python
terr['Wounded'] = terr['Wounded'].astype('int64')
terr['Killed'] = terr['Killed'].astype('int64')
```

```python
terr.info()
```
