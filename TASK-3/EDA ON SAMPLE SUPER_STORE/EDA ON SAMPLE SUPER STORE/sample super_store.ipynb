## uni-variant analysis

### UNI-VARIANT ANALYSIS ON REGION

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Region',data=super_store)
plt.show()
```

- West as most sales.
- south as least sales.

### UNI-VARIANT ANALYSIS ON CATEGORY.

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Category',data=super_store)
plt.show()
```

- Office supplies category as most sales.

### UNI-VARIANT ANALYSIS ON SUB-CATEGORY.

```python
plt.figure(figsize=(17,10))
sns.countplot(x='Sub-Category',data=super_store)
plt.show()
```

- Binders and Paper are most of the sales in the sub-category.

### UNI-VARIANT ANALYSIS ON SEGMENT.

```python
plt.figure(figsize=(5,5))
sns.countplot(x='Segment',data=super_store)
plt.show()
```

- Consumer are most of the sale in the segment.

### UNI-VARIANT ANALYSIS ON SALES.

```python
plt.figure(figsize=(10,5))
sns.kdeplot(super_store['Sales'],color='black')
plt.title('SALES')
plt.show()
```

- MOST OF THE SALES ARE RANGES IN BETWEEN 0 TO 5000.
- FROM 5000 SALES ARE MOST SKEWED.

### UNI-VARIANT ANALYSIS ON PROFIT.

```python
plt.figure(figsize=(10,5))
sns.kdeplot(super_store['Profit'])
plt.title('PROFIT')
plt.show()
```

- Profits are less in the sales.

 ## BI-VARIANT ANALYSIS

### COMPARING SUB-CATEGORY AND SALES.

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='Sub-Category',y='Sales',data=super_store)
plt.xticks(rotation=90,fontsize=10)
plt.title('COMPARING THE SUB-CATEGORY WITH SALES')
plt.show()
```

- MACHINES AND COPIERS ARE MOST OF THE SALES IN THE SUB-CATEGORY.

### COMPARING SALES AND PROFIT & REGION.

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='Sales',y='Profit',data=super_store,hue='Region')
plt.xticks(rotation=90,fontsize=10)
plt.title('COMPARING SALES AND PROFIT & REGION.')
plt.show()
```

- SALES AND PROFIT ARE MORE RELATED.
- WEST AND CENTRAL HAVE MORE SALES AND PROFIT.

### COMPARING CATEGORY AND PROFIT & REGION.

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='Category',y='Profit',data=super_store,hue='Region')
plt.xticks(rotation=90,fontsize=10)
plt.title('COMPARING CATEGORY AND PROFIT & REGION.')
plt.show()
```

- TECHNOLOGY HAVING MORE PROFIT.
- FURNITURE HAVING LESS PROFIT.
- MOST OF THE PROFITS ARE FROM CENTRAL AND WEST.

### COMPARING STATE AND SALES.

```python
plt.figure(figsize=(50,30))
sns.barplot(x=super_store['State'],y=super_store['Sales'])
plt.title('COMPARING STATE AND SALES.')
plt.show()
```

- VERMONT AND WYOMING HAVING MORE SALES IN THE STATES.

### COMPARING STATE AND PROFIT.

```python
plt.figure(figsize=(60,30))
sns.barplot(x=super_store['State'],y=super_store['Profit'])
plt.title('COMPARING STATE AND PROFIT.')
plt.show()
```

- MONTANA AND VERMONT HAVING MORE PROFITS IN THE SALES.

### COMPARING CATEGORY AND SEGMENT.

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Category',data=super_store,hue='Segment',palette="rocket_r")
plt.xticks(rotation=90,fontsize=10)
plt.title('COMPARING CATEGORY AND SEGMENT')
plt.show()
```

- IN SEGMENT OFFICE SUPPLIES ARE MORE.

### COMPARING CATEGORY AND REGION.

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Category',data=super_store,hue='Region',palette="rocket_r")
plt.xticks(rotation=90,fontsize=10)
plt.title('COMPARING CATEGORY AND REGION.')
plt.show()
```

- IN REGION OFFICE SUPPLIES ARE MORE.

### COMPARING SALES AND PROFIT & DISCOUNT.

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='Sales',y='Profit',data=super_store,hue='Discount')
plt.xticks(rotation=90,fontsize=10)
plt.title('COMPARING SALES AND PROFIT & DISCOUNT.')
plt.show()
```

- DISCOUNT ARE LESS FOR SALES.
- FOR HIGH DISCOUNT PROFITS ARE LESS.

### COMPARING SUB-CATEGORY AND PROFIT & DISCOUNT.

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='Sub-Category',y='Profit',data=super_store,hue='Discount')
plt.xticks(rotation=90,fontsize=10)
plt.title('COMPARING SUB-CATEGORY AND PROFIT & DISCOUNT.')
plt.show()
```

- BLINDERS AND MACHINES SALES ARE GIVING DISCOUNT.
- BLINDERS AND MACHINES PROFITS ARE LESS AND THERE IS IN LOSS.

```python
sns.pairplot(super_store,hue='Sub-Category')
plt.show()
```
